```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Bidirectional, GRU, Dropout, Attention
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from gensim.models import Word2Vec
import warnings
warnings.filterwarnings('ignore')

# Excel veri setini yükleme
dataset_path = 'combined_last_df.xlsx'  # Excel dosya yolunu belirtin
try:
    # Excel dosyasını oku
    df = pd.read_excel(dataset_path)
    print("Excel veri seti başarıyla yüklendi!")
    print(f"Veri seti boyutu: {df.shape}")

    # İlk birkaç satırı görüntüle
    print("\nVeri setinden ilk 5 satır:")
    print(df.head())

    # Sütun isimlerini görüntüle
    print("\nSütun isimleri:")
    print(df.columns.tolist())

    # Tweet sütunu - ekran görüntüsünden "Tweet" adında bir sütun olduğu görülüyor
    tweets = df['Tweet'].values

    # Etiket sütunu - ekran görüntüsünden "Etiket" adında bir sütun olduğu görülüyor
    # Burada birden fazla etiket sütunu var gibi görünüyor, en uygun olanı seçin
    # Örneğin: "Etiket" veya "Etiket Alt" sütunu

    # İkili sınıflandırma için: "nefret" içeren her şey 1, diğerleri 0 olarak etiketlenebilir
    labels = np.zeros(len(df))
    for i, etiket in enumerate(df['Son Etiket Etiket'].values):
        if isinstance(etiket, str) and 'nefret' in etiket.lower():
            labels[i] = 1

    print(f"\nToplam tweet sayısı: {len(tweets)}")
    print(f"Nefret söylemi içeren tweet sayısı: {int(sum(labels))}")
    print(f"Normal tweet sayısı: {len(labels) - int(sum(labels))}")

except Exception as e:
    print(f"Excel dosyası yükleme hatası: {e}")
    print("Lütfen Excel dosya yolunu ve sütun isimlerini kontrol edin.")
    # Hata olursa işlemi durdurmak için
    import sys
    sys.exit(1)

# Verileri ön işleme
print("\nVeriler ön işleniyor...")

# Tweet'leri temizleme (özel karakterler, linkler vb.)
def clean_text(text):
    if isinstance(text, str):
        # URL'leri kaldır
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        # HTML etiketlerini kaldır
        text = re.sub(r'<.*?>', '', text)
        # Özel karakterleri kaldır
        text = re.sub(r'[^\w\s]', '', text)
        # Fazla boşlukları kaldır
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    return ""

# Tweet'leri temizle
cleaned_tweets = np.array([clean_text(tweet) for tweet in tweets])

# Boş tweet'leri temizle
valid_indices = [i for i, tweet in enumerate(cleaned_tweets) if len(tweet.strip()) > 0]
cleaned_tweets = cleaned_tweets[valid_indices]
cleaned_labels = labels[valid_indices]

print(f"Temizleme sonrası kalan tweet sayısı: {len(cleaned_tweets)}")

# Veriyi eğitim, doğrulama ve test setlerine ayırma
X_train_val, X_test, y_train_val, y_test = train_test_split(
    cleaned_tweets, cleaned_labels, test_size=0.2, random_state=42, stratify=cleaned_labels
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
)

print(f"Eğitim seti: {len(X_train)} tweet")
print(f"Doğrulama seti: {len(X_val)} tweet")
print(f"Test seti: {len(X_test)} tweet")

# Word2Vec için tweetleri tokenize etme
tokenized_tweets = [tweet.split() for tweet in cleaned_tweets]

# 1. Turkish Word2Vec modelini ince ayarlama
print("\nWord2Vec modeli yükleniyor ve fine-tune ediliyor...")
try:
    # Var olan modeli yükleme
    word2vec_model = Word2Vec.load("trword2vec_finetuned_local.model")
    print("Mevcut Word2Vec modeli yüklendi.")
    # Mevcut model üzerinde fine-tuning yapma
    word2vec_model.train(tokenized_tweets, total_examples=len(tokenized_tweets), epochs=5)
except:
    print("Mevcut Word2Vec modeli bulunamadı, yeni bir model eğitiliyor...")
    # Yeni bir model eğitme
    word2vec_model = Word2Vec(tokenized_tweets, vector_size=300, window=5, min_count=1, workers=4)

# Word2Vec modelini kaydetme
word2vec_model.save("fine_tuned_turkish_word2vec.model")
print("Word2Vec modeli fine-tune edildi ve kaydedildi.")

# Tokenizer oluşturma ve fit etme
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
vocab_size = len(tokenizer.word_index) + 1
print(f"Kelime haznesi boyutu: {vocab_size}")

# Maksimum tweet uzunluğunu belirleme
max_length = max([len(tweet.split()) for tweet in cleaned_tweets])
print(f"Maksimum tweet uzunluğu: {max_length}")

# Tweetleri sayısal dizilere dönüştürme
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Tüm dizileri aynı uzunluğa padding ile getirme
X_train_pad = pad_sequences(X_train_seq, maxlen=max_length, padding='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=max_length, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_length, padding='post')

# Word2Vec embeddingi için matris oluşturma
embedding_matrix = np.zeros((vocab_size, 300))
for word, i in tokenizer.word_index.items():
    try:
        embedding_vector = word2vec_model.wv[word]
        embedding_matrix[i] = embedding_vector
    except KeyError:
        continue

# 2-3. Encoder-Decoder Model ve Attention Mekanizmaları
def create_encoder_decoder_attention(cell_type='lstm', attention_type='bahdanau', use_pretrained=True):
    """
    Encoder-Decoder modeli oluşturan fonksiyon

    Parametreler:
    cell_type: RNN hücresi tipi ('lstm', 'bilstm', 'gru')
    attention_type: Dikkat mekanizması tipi ('bahdanau', 'luong', 'scaled_dot')
    use_pretrained: Word2Vec ağırlıklarını kullanıp kullanmama

    Dönüş:
    Oluşturulan model
    """
    # Input katmanı
    inputs = Input(shape=(max_length,))

    # Embedding katmanı
    if use_pretrained:
        embedding_layer = Embedding(
            vocab_size,
            300,
            weights=[embedding_matrix],
            input_length=max_length,
            trainable=False
        )
    else:
        embedding_layer = Embedding(
            vocab_size,
            300,
            input_length=max_length
        )

    x = embedding_layer(inputs)

    # Encoder katmanı
    if cell_type == 'lstm':
        encoder = LSTM(128, return_sequences=True)(x)
    elif cell_type == 'bilstm':
        encoder = Bidirectional(LSTM(128, return_sequences=True))(x)
    elif cell_type == 'gru':
        encoder = GRU(128, return_sequences=True)(x)

    # Attention mekanizması
    if attention_type == 'bahdanau':
        # Bahdanau (additive) attention
        attention_layer = tf.keras.layers.Attention()([encoder, encoder])
    elif attention_type == 'luong':
        # Luong (dot-product) attention
        attention_layer = tf.keras.layers.Attention(use_scale=False)([encoder, encoder])
    elif attention_type == 'scaled_dot':
        # Scaled dot-product attention
        attention_layer = tf.keras.layers.Attention(use_scale=True)([encoder, encoder])

    # Pooling katmanı
    pooled = tf.keras.layers.GlobalAveragePooling1D()(attention_layer)

    # Dropout katmanı
    dropout = Dropout(0.5)(pooled)

    # Çıkış katmanı (ikili sınıflandırma için sigmoid aktivasyon)
    outputs = Dense(1, activation='sigmoid')(dropout)

    # Model oluşturma
    model = Model(inputs=inputs, outputs=outputs)

    # Model derleme
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# Modelleri eğitip değerlendirme için fonksiyon
def train_and_evaluate_model(cell_type, attention_type, use_pretrained, X_train, y_train, X_val, y_val, X_test, y_test):
    """
    Modeli eğitip değerlendiren fonksiyon
    """
    # Model adını oluşturma
    model_name = f"{cell_type}_{attention_type}_{'pretrained' if use_pretrained else 'random'}"
    print(f"\nModel eğitiliyor: {model_name}")

    # Eğitim başlama zamanı
    train_start_time = time.time()

    # Modeli oluşturma
    model = create_encoder_decoder_attention(cell_type, attention_type, use_pretrained)

    # Modelin özeti
    print("\nModel özeti:")
    model.summary()

    # Erken durdurma ve model checkpoint
    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(f"{model_name}.h5", monitor='val_loss', save_best_only=True)

    # Modeli eğitme
    history = model.fit(
        X_train_pad, y_train,
        validation_data=(X_val_pad, y_val),
        epochs=15,
        batch_size=64,
        callbacks=[early_stopping, model_checkpoint]
    )

    # Eğitim süresi
    train_time = time.time() - train_start_time

    # Tahmin başlama zamanı
    predict_start_time = time.time()

    # Test seti üzerinde tahmin
    y_pred_prob = model.predict(X_test_pad)
    y_pred = (y_pred_prob > 0.5).astype(int).flatten()

    # Tahmin süresi
    predict_time = time.time() - predict_start_time

    # Değerlendirme metrikleri
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Sonuçları raporlama
    print(f"\nModel: {model_name}")
    print(f"Eğitim süresi: {train_time:.2f} saniye")
    print(f"Tahmin süresi: {predict_time:.2f} saniye")
    print(f"Accuracy: {np.mean(y_pred == y_test):.3f}")
    print(f"Precision: {precision:.3f}")
    print(f"Recall: {recall:.3f}")
    print(f"F1 Score: {f1:.3f}")

    # Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Nefret Söylemi'],
                yticklabels=['Normal', 'Nefret Söylemi'])
    plt.xlabel('Tahmin')
    plt.ylabel('Gerçek')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.savefig(f"confusion_matrix_{model_name}.png")

    # Eğitim ve doğrulama metrikleri
    plt.figure(figsize=(12, 5))

    # Loss grafik
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Eğitim Loss')
    plt.plot(history.history['val_loss'], label='Doğrulama Loss')
    plt.title(f'Loss - {model_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Accuracy grafik
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Eğitim Accuracy')
    plt.plot(history.history['val_accuracy'], label='Doğrulama Accuracy')
    plt.title(f'Accuracy - {model_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig(f"training_metrics_{model_name}.png")

    return {
        'model_name': model_name,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'accuracy': np.mean(y_pred == y_test),
        'train_time': train_time,
        'predict_time': predict_time,
        'confusion_matrix': conf_matrix,
        'history': history.history
    }

# Ana program - modelleri eğitme ve değerlendirme
print("\nModel eğitimi başlatılıyor...")

# Tüm model kombinasyonları
cell_types = ['lstm', 'bilstm', 'gru']
attention_types = ['bahdanau', 'luong', 'scaled_dot']
embedding_types = [True, False]  # True: pretrained, False: random

# Test için sadece bir kombinasyon (daha hızlı sonuç için)
# Tüm kombinasyonlar için aşağıdaki satırları yorumdan çıkarın
# cell_types = ['lstm']
# attention_types = ['bahdanau']
# embedding_types = [True]

results = []

for cell_type in cell_types:
    for attention_type in attention_types:
        for use_pretrained in embedding_types:
            result = train_and_evaluate_model(
                cell_type, attention_type, use_pretrained,
                X_train_pad, y_train, X_val_pad, y_val, X_test_pad, y_test
            )
            results.append(result)

# Sonuçları DataFrame'e dönüştürme
results_df = pd.DataFrame([
    {
        'Model': r['model_name'],
        'Accuracy': r['accuracy'],
        'Precision': r['precision'],
        'Recall': r['recall'],
        'F1': r['f1'],
        'Training Time (s)': r['train_time'],
        'Prediction Time (s)': r['predict_time']
    } for r in results
])

# Sonuçları CSV olarak kaydetme
results_df.to_csv('model_comparison_results.csv', index=False)

# En iyi modeli belirleme
best_model_idx = results_df['F1'].idxmax()
best_model = results_df.iloc[best_model_idx]

print("\n--- EN İYİ MODEL ---")
print(f"Model: {best_model['Model']}")
print(f"Accuracy: {best_model['Accuracy']:.3f}")
print(f"Precision: {best_model['Precision']:.3f}")
print(f"Recall: {best_model['Recall']:.3f}")
print(f"F1 Score: {best_model['F1']:.3f}")
print(f"Eğitim Süresi: {best_model['Training Time (s)']:.2f} saniye")
print(f"Tahmin Süresi: {best_model['Prediction Time (s)']:.2f} saniye")

# Tüm sonuçları gösteren ısı haritası
plt.figure(figsize=(12, 8))
sns.heatmap(
    results_df[['Accuracy', 'Precision', 'Recall', 'F1']].values,
    annot=True,
    fmt='.3f',
    cmap='viridis',
    xticklabels=['Accuracy', 'Precision', 'Recall', 'F1'],
    yticklabels=results_df['Model']
)
plt.title('Model Performans Karşılaştırması')
plt.tight_layout()
plt.savefig('model_performance_comparison.png')

print("\nİşlem tamamlandı. Detaylı sonuçlar 'model_comparison_results.csv' dosyasında.")
```
